## Testing speed of list and numpy arrays

```python
from time import process_time
```

```python
# LIST
py_list = [i for i in range(10000)]

start_time = process_time()

py_list = [i+5 for i in py_list]

end_time = process_time()

print(end_time - start_time)
```

```python
import numpy as np
```

```python
# NUMPY Array
np_array = np.array([i for i in range(10000)])

start_time = process_time()

np_array += 5

end_time = process_time()


print(end_time - start_time)
```

```python
# numpy array dimension

np_array = np.array([(1,2,3),(4,4,6)])
np_array.shape
```

```python
#  set a data type
a = np.array([(1,2,3), (4,5,6)], dtype = float)
print(a)
```

```python
# Placeholders - 0s

np_zeros = np.zeros((3,3))
print(np_zeros)
```

```python
# Placeholder - 1s

np_ones = np.ones((3,2))
print(np_ones)
```

```python
# Placeholder - Full

np_full = np.full((4,2), 5)
print(np_full)
```

```python
# Identity Matrix

i = np.eye(4)
print(i)
```

```python
# Random Variable -> Default is float

r = np.random.random((2,2))
print(r)
```

```python
# Rand Int within a range of 10 and 20 ===> matrix dim: 2x3

r = np.random.randint(10,20,(2,3))
print(r)
```
